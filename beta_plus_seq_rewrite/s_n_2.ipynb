```python
def check_seq(n, bs, inv=False):
    # gen formula (s2-v)%s1
    min_s1 = max([bs[0]] + bs[n+1:])
    m = 2*n+1
    for s1 in range(min_s1, m+1):
        for s2 in range(0, s1):
            is_compat = True
            seq_used = [False] * m
            seq_bs = [None] * m
            for idx in range(m):
                seq_bs[idx] = bs[idx]
                if (idx == 0 or idx >= n+1) != inv:
                    seq_bs[idx] = (s2-bs[idx])%s1                    
                if seq_used[seq_bs[idx]]:
                    is_compat = False
                    break
                seq_used[seq_bs[idx]] = True
            if not is_compat:
                continue
            sums = set()
            for idx in range(n):
                e_sum = seq_bs[0] + seq_bs[idx + 1]
                if e_sum in sums:
                    is_compat = False
                    break
                sums.add(e_sum)
            for idx in range(n):
                e_sum = seq_bs[idx + 1] + seq_bs[n+1+idx]
                if e_sum in sums:
                    is_compat = False
                    break
                sums.add(e_sum)
            if max(sums)-min(sums)+1 != 2*n:
                is_compat = False
            if not is_compat:
                continue
            if not inv:
                print(inv, bs, seq_bs, s1, s2)
            else:
                print(inv, [2*n-x for x in bs], [2*n-x for x in seq_bs], s1, s2)


def solve3(n, bs, used, has_diffs, idx):
    if idx == n:
        check_seq(n, bs)
        inv_bs = [2*n - x for x in bs]
        check_seq(n, inv_bs, inv=True)
        return
    for v in range(0, bs[idx+1]):
        if used[v]:
            continue
        diff = bs[idx+1] - v
        if has_diffs[diff]:
            continue
        bs[n+1+idx] = v
        used[v] = True
        has_diffs[diff] = True
        solve3(n, bs, used, has_diffs, idx+1)
        used[v] = False
        has_diffs[diff] = False

def solve2(n, bs, used, has_diffs, idx):
    if idx == n:
        solve3(n, bs, used, has_diffs, 0)
        return
    if idx == 0:
        last_v = 2*n+1
    else:
        last_v = bs[idx]
    for v in range(last_v-1, 1, -1):
        if used[v]:
            continue
#         if n % 3 == 0 and v % 4 == (2*n-2) % 4:
#             continue
        if v < bs[0]:
            return
        diff = v - bs[0]
        if has_diffs[diff]:
            continue
        bs[idx + 1] = v
        used[v] = True
        has_diffs[diff] = True
        solve2(n, bs, used, has_diffs, idx+1)
        used[v] = False
        has_diffs[diff] = False

def solve(n):
    bs = [None] * (2*n+1)
    used = [False] * (2*n+1)
    has_diffs = [False] * (2*n+1)
    for i in range(0, n+1):
        bs[0] = i
        used[i] = True
        solve2(n, bs, used, has_diffs, 0)
        used[i] = False
```
